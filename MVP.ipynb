```python
import pandas as pd
import requests
import json
import csv
import os

from sqlalchemy import create_engine

# Output File (CSV)
output_data_file = "Resources/clubs.csv"
```

```python
# Which country has the most wins? And creating dataframe with results
world = "Resources/WorldCups.csv"
world_df = pd.read_csv(world)
most_wins = world_df["Winner"].value_counts()
most_wins_df = pd.DataFrame(most_wins)

# Resetting index to keep value counts results as varables
# Renameing index column
most_wins_df = most_wins_df.reset_index()
most_wins_df.rename(columns={"index": "Country"})
```

```python
url = pd.read_html("https://en.wikipedia.org/wiki/FIFA_Club_World_Cup")
clubs_country = url[3]
# clubs_country = pd.DataFrame(clubs_country) # Alap: This is not needed since read_html gives you DataFrame object.
```

```python
with open(output_data_file, 'w') as csv_file:
    clubs_country.to_csv(csv_file, index=False)
```
